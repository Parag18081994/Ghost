```python
#Categories (2, object): [female, male]-it will give unique values(Mlae,Female)
new_sex=pd.Categorical(titanic_train['Sex'])
new_sex
```

here we are extracting the first letter of each entry

```python
#here we are extracting the first letter of each entry and find categorical values i.e 8
titanic_train['Cabin']
char_Cabin=titanic_train['Cabin'].astype(str)#convert to str for list comprehension operation because by deafult it is in object format
new_cabin=[i[0] for i in char_Cabin]
new_cabin=pd.Categorical(new_cabin)
new_cabin
```

```python
#here we are creating a new col-cabin_1
titanic_train['cabin_1']=new_cabin
titanic_train
```

```python
#if want to overwrite
titanic_train['Cabin']=new_cabin
titanic_train
```

```python
#Below is implementation of above cell-46 line-4 of list comprehension
l=[]
for i in char_Cabin:
    new_cabin=i[0]
    l.append(new_cabin)
l
pd.DataFrame(l)
```
